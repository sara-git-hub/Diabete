### Apprentissage non supervisee

Import du jeu de donnees

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Importation des donnees 
X = np.array([
    [15, 1],
    [16, 2],
    [14, 1.5],
    [60, 6],
    [65, 5],
    [62, 7],
    [120, 11],
    [115, 10],
    [122, 12]
])
```

Visualisation de la donnee

```python
plt.scatter(X[:, 0], X[:, 1], s=100)
plt.xlabel("Revenues(en milliers euros)")
plt.ylabel("Frequence d'achat par (mois)")
plt.title("Client sans etiquettes")
plt.grid(True)
plt.show()
```

Application du KMeans avec trois clusters

```python
kmeans = KMeans(n_clusters= 3, random_state= 42)
kmeans.fit(X)

labels = kmeans.labels_
print(labels)
```

Visualisation des clusters trouves

```python
centers = kmeans.cluster_centers_

print(centers)
plt.scatter(X[:, 0], X[:, 1], c = labels, cmap= "viridis", s=100)
plt.scatter(centers[:, 0], centers[:, 1], c= 'red', s= 200, marker= "X", label = 'Centres')
plt.xlabel("Revenu (En millier d'Euro)")
plt.ylabel("Frequence d'achat par mois")
plt.title("Clusters detectes par KMeans")
plt.legend()
plt.grid(True)
plt.show()
```
